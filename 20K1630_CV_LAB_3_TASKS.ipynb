```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

gray = cv2.imread('pic7.png', cv2.IMREAD_GRAYSCALE)
rgb = cv2.imread('pic6.png')

resize = cv2.resize(gray, (rgb.shape[1], rgb.shape[0]))

colormaps = [
    cv2.COLORMAP_AUTUMN,
    cv2.COLORMAP_BONE,
    cv2.COLORMAP_JET,
    cv2.COLORMAP_WINTER,
    cv2.COLORMAP_RAINBOW,
    cv2.COLORMAP_OCEAN,
    cv2.COLORMAP_SUMMER,
    cv2.COLORMAP_SPRING,
    cv2.COLORMAP_COOL,
    cv2.COLORMAP_HSV,
    cv2.COLORMAP_PINK,
    cv2.COLORMAP_HOT,
]

gray_res = []
rgb_res = []

for colormap in colormaps:
    gray_result = cv2.applyColorMap(resize, colormap)
    gray_res.append(gray_result)

    rgb_result = cv2.applyColorMap(rgb, colormap)
    rgb_res.append(rgb_result)

gray_disp = np.hstack(gray_res[:4]), np.hstack(gray_res[4:8]), np.hstack(gray_res[8:])
rgb_disp = np.hstack(rgb_res[:4]), np.hstack(rgb_res[4:8]), np.hstack(rgb_res[8:])

cv2_imshow(np.vstack(gray_disp))
cv2_imshow(np.vstack(rgb_disp))
```

```python
save_gray = np.hstack(gray_res[:4]), np.hstack(gray_res[4:8]), np.hstack(gray_res[8:])
save_color = np.hstack(rgb_res[:4]), np.hstack(rgb_res[4:8]), np.hstack(rgb_res[8:])
```

```python
cv2.imwrite('saved_gray.png', np.vstack(save_gray))
cv2.imwrite('saved_color.png', np.vstack(save_color))
```
